```python
# Import Dependencies
import pandas as pd
import numpy as np
import pathlib as pl
```

```python
# Read the data
national_od = pd.read_csv('Resources/VSRR_Provisional_Drug_Overdose_Death_Counts.csv')
national_od.head()
```

```python
national_od.info()
```

```python
# Get only desired columns
national_od = national_od.loc[:, ["State", "Year", "Indicator", "Data Value", "State Name"]]
national_od
```

```python
national_od = national_od.loc[national_od["Indicator"] == "Number of Drug Overdose Deaths"]
national_od
```

```python
national_od = national_od.replace(',','', regex=True)
```

```python
national_od["Data Value"] = national_od["Data Value"].astype(str).astype(int)
```

```python
national_od.info()
```

```python
national_od = national_od.groupby(["State", "Year"])
national_od["Data Value"].sum()
```
