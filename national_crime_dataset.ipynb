```python
# Import Dependencies
import pandas as pd
import numpy as np
import pathlib as pl
```

```python
# Read the data
national_crime = pd.read_csv('Resources/state_crime.csv')
national_crime.head()
```

```python
national_crime.info()
```

```python
# Get only desired columns
national_crime = national_crime.loc[:, ["State", "Year", "Data.Totals.Property.All", "Data.Totals.Violent.All"]]
national_crime
```

```python
national_crime = national_crime.loc[national_crime["Year"] >= 2015]
national_crime
```

```python
national_crime = national_crime.replace(',','', regex=True)
```

```python
national_crime["Data.Totals.Property.All"] = national_crime["Data.Totals.Property.All"].astype(str).astype(int)
national_crime["Data.Totals.Violent.All"] = national_crime["Data.Totals.Violent.All"].astype(str).astype(int)
```

```python
national_crime.info()
```

```python
national_crime['Crime Data'] = national_crime[["Data.Totals.Property.All", "Data.Totals.Violent.All"]].sum(axis=1)
national_crime
```

```python
national_crime = pd.DataFrame(national_crime.groupby(["State", "Year"])["Crime Data"].sum())
national_crime
```

```python
file_name = 'crime_data.xlsx'
national_crime.to_excel(file_name)
```
