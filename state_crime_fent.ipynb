```python
columns_to_keep= ['State','Year', 'Data.Population',"Data.Rates.Violent.All"]
crime= pd.read_csv('state_crime.csv', usecols=columns_to_keep)
```

```python
# checking for missing values
crime.isnull().sum()
```

```python
# checking for duplicates
crime.duplicated().sum()
```

```python
# filtering to only keep the year between 2015-2019
years_to_keep= list(range(2015,2019))
crime_filtered = crime[crime['Year'].isin(years_to_keep)]
```

```python
years_to_drop= [str(year) for year in range(1960, 2019)]
crime_filtered = crime.drop(columns=[col for col in crime.columns if col in years_to_drop])
print(crime_filtered)
```

```python
crime_filtered.to_csv('filtered_crime_data.csv', index=False)
```
